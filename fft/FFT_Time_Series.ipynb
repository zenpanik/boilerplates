```python
def exp_ma_func(x,window):
    exp_ma = []
    N = len(x)
    for i in range(N-len(window)):
        exp_ma.append(sum(np.array(x[i+len(window):i:-1])*window))
    return exp_ma
```

```python
window = np.array([0.5, 0.5**2, 0.5**3, 0.5**4])
window = window.conj().transpose()
window = window/sum(window)
```

```python
window
```

```python
df_work['150_ma'] = np.nan
```

```python
df_work.loc[len(window):, '150_ma'] = exp_ma_func(df_work['150'],window)
```

```python
x_org_std = float(np.std(df_work['150']))
```

```python
df_work['150_ma_cap'] = df_work['150_ma'] + 1.96*x_org_std
df_work['150_ma_floor'] = df_work['150_ma'] - 1.96*x_org_std
```

```python
df_work.plot()
```

```python
max_val = np.max(df_work['150'])
std_val = np.std(df_work['150'])

x_norm = df_work['150']/max_val
x = x_norm

n = x.size

t = np.arange(0, n)
p = np.polyfit(t, x, 1)         # find linear trend in x
z = np.poly1d(p)
```

```python
x_notrend = x - p[0] * t[::]
```

```python
x_notrend.plot()
```

```python
from numpy import fft
n = x.size
x_freqdom = fft.fft(x_notrend)  # detrended x in frequency domain
DC = np.absolute(x_freqdom[0])
x_freqdom[0] = 0.0
f = fft.fftfreq(n)              # frequencies
```

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.dates as mdates
import matplotlib.patches as mpatches
fig_size = (16, 9)
font = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 32}

matplotlib.rc('font', **font)
```

```python
# Visualize Power Spectra Density of the train data set
# Visually select meaningful frequencies for seasonal components evaluation - n_harm in next cell

fig = plt.figure(figsize = fig_size)
ax = fig.add_subplot(111)

fig.suptitle("Price in frequency domain")
ax.set_ylabel("Amplitude")
ax.set_xlabel("Normalized frequency")

ax = fig.add_subplot(111)
plt.scatter(f[:int(n/2):], np.absolute(x_freqdom[:int(n/2):]), c="#393ACC")
ax.add_patch(
    mpatches.Rectangle(
        (0.0, 7.5),
        0.5,
        90,
        fill=False,      # remove background
        lw=2.0,
        edgecolor='#00ff55'
    )
    )
ax.xaxis.label.set_fontsize(28)
ax.yaxis.label.set_fontsize(28)
for label in (ax.get_xticklabels() + ax.get_yticklabels()):
    label.set_fontsize(22)
    
plt.savefig("train_data_Freq_components")
fig.show()
```

```python
x_freqdom[0] = 0.0
n_predict = len(x)
n_harm = 100
indexes = sorted(np.arange(int(len(x_freqdom)/2)), key=lambda i: x_freqdom[i], reverse=True)
t = np.arange(0, n)
restored_sig = np.zeros(t.size)
for i in indexes[0:n_harm]:
    ampli = np.absolute(x_freqdom[i])  # amplitude
    phase = np.angle(x_freqdom[i])              # phase
    restored_sig += ampli * np.cos(2 * np.pi * f[i] * t + phase)
```

```python
predicted_trend_season = (restored_sig + z(t)*max_val)
```

```python
df_work['150_predicted'] = predicted_trend_season
```

```python
df_work['150_ifft'] = restored_sig
```

```python
df_work[['150_predicted', '150']].plot(legend=False)
```

```python
detrended = (np.array(x[0]) - predicted_trend_season[:-n_predict:])[n_harm::]
```

```python
df_work['detrended'] = df_work['150'] - df_work['150_predicted']
```

```python
detrended = df_work['detrended'].values
```

```python
# Create Factors matrix for OLS regression 

x_in = np.column_stack((detrended[5:-1:], detrended[4:-2:], detrended[3:-3:], detrended[2:-4:], detrended[1:-5:], detrended[:-6:]))
```

```python
import statsmodels.regression.linear_model as sm
```

```python
# Fit OLS regression model to detrended train data
# Print results from model fitting
olsmod = sm.OLS(detrended[6::], x_in)
olsres = olsmod.fit()
regr_fit_summary = olsres.summary()
print(olsres.summary())
```
